# Clearing time descrepencies

```python
import MetaTrader5 as mt5
import pandas as pd
%matplotlib nbagg
```

```python
mt5.initialize()
```

```python
usdrub = pd.DataFrame(mt5.copy_rates_from_pos('USDRUB', mt5.TIMEFRAME_H1, 0, 48)).drop(columns=['spread', 'real_volume'])
audusd = pd.DataFrame(mt5.copy_rates_from_pos('AUDUSD', mt5.TIMEFRAME_H1, 0, 48)).drop(columns=['spread', 'real_volume'])
usdrub['time'] = pd.to_datetime(usdrub['time'], unit='s')
audusd['time'] = pd.to_datetime(audusd['time'], unit='s')
```

```python
usdrub.tail(12)
```

```python
audusd.tail(12)
```

```python
usdrub.head(12)
```

```python
audusd.head(12)
```

```python
audusd.set_index(keys=['time'], inplace=True)
usdrub.set_index(keys=['time'], inplace=True)
```

```python
audrub = audusd[['open', 'high', 'low', 'close']] * usdrub[['open', 'high', 'low', 'close']]
audrub.head(24)
```

```python
audrub.tail(48)
```

```python
audrub.dropna(how='all', inplace=True)
audrub.tail(48)
```

```python
audrub.plot(legend=True)
```

```python
audrub.dropna(how='all', inplace=True)
```
