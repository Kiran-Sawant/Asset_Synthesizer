# non-dollar non-forex / dollar-quote forex
### DAX AUDUSD

we need to extract the data from MT5 after determining the nature of asset and currency.

```python
def dollarPair(currency: str):
    if currency in ['EUR', 'GBP', 'AUD', 'NZD']:
        return currency + 'USD'
    else:
        return 'USD' + currency
```

```python
profitCurrency = mt5.symbol_info('DE30').currency_profit
profitCurrency
```

```python
def dollarizer(asset, dollarpair):
    
    asset_rates = pd.DataFrame(mt5.copy_rates_from_pos(asset, mt5.TIMEFRAME_H1, 0, 10)).drop(columns=['spread', 'real_volume'])
    dollar_rates = pd.DataFrame(mt5.copy_rates_from_pos(dollarpair, mt5.TIMEFRAME_H1, 0, 10)).drop(columns=['spread', 'real_volume'])
        
    asset_rates['time'] = pd.to_datetime(asset_rates['time'], unit='s')
    dollar_rates['time'] = pd.to_datetime(dollar_rates['time'], unit='s')
        
    asset_rates.set_index(keys=['time'], inplace=True)
    dollar_rates.set_index(keys=['time'], inplace=True)
    
    if dollarpair[3:6] == 'USD':     # quoted in dollar
        dollarised_asset = asset_rates[['open', 'high', 'low', 'close']] * dollar_rates[['open', 'high', 'low', 'close']]
    
    else:                             # based in dollar
        dollarised_asset = asset_rates[['open', 'high', 'low', 'close']] / dollar_rates[['open', 'high', 'low', 'close']]
        dollarised_asset.rename(columns={'high': 'low', 'low': 'high'})
    
    dollarised_asset['vol'] = (asset_rates['tick_volume'] + dollar_rates['tick_volume']) / 2
    return dollarised_asset
        
```

```python
dollarizer("DE30", 'EURUSD').round(2)
```
